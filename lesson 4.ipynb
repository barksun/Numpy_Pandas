### Задание 1
#### Импортируйте библиотеки pandas, numpy и matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

#### Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn.  

```python
from sklearn.datasets import load_boston
```

#### Создайте датафреймы X и y из этих данных.

```python
boston = load_boston()

boston.keys()
```

```python
data = boston["data"]

data
```

```python
feature_names = boston["feature_names"]

feature_names
```

```python
target = boston["target"]

target
```

```python
X = pd.DataFrame(data, columns=feature_names)

X.info()
```

```python
X.head()
```

```python
y = pd.DataFrame(target, columns=["price"])

y.info()
```

```python
y.head()
```

#### Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 20% от всех данных, при этом аргумент random_state должен быть равен 42.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

#### Масштабируйте данные с помощью StandardScaler.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_stand_features = scaler.fit_transform(X_train)
```

```python
X_train_stand_features
```

```python
X_train_stand_features.shape
```

```python
X_train_scaled = pd.DataFrame(X_train_stand_features, columns=feature_names)
```

```python
X_train_scaled.head()
```

```python
X_test_stand_features = scaler.fit_transform(X_test)
```

```python
X_test_stand_features
```

```python
X_test_stand_features.shape
```

```python
X_test_scaled = pd.DataFrame(X_test_stand_features, columns=feature_names)
```

```python
X_test_scaled.head()
```

#### Постройте модель TSNE на тренировочный данных с параметрами: n_components=2, learning_rate=250, random_state=42.

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, learning_rate=250, random_state=42)

X_train_tsne = tsne.fit_transform(X_train_scaled)

print('До:\t{}'.format(X_train_scaled.shape))
print('После:\t{}'.format(X_train_tsne.shape))
```

#### Постройте диаграмму рассеяния на этих данных.

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1])

plt.show()
```

### Задание 2
#### С помощью KMeans разбейте данные из тренировочного набора на 3 кластера, используйте все признаки из датафрейма X_train. Параметр max_iter должен быть равен 100, random_state сделайте равным 42.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, max_iter=100, random_state=42)

labels_train = kmeans.fit_predict(X_train_scaled)

plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1], c=labels_train)

plt.show()
```

#### Постройте еще раз диаграмму рассеяния на данных, полученных с помощью TSNE,и раскрасьте точки из разных кластеров разными цветами.

```python
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1])

plt.show()
```
